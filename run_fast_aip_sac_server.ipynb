# Fast AIP SAC training

SAC 학습을 Jupyter에서 실행한다. 학습을 멈출 때는 커널을 종료하지 말고 먼저 **Interrupt Kernel**을 사용한다.

```python
import json
import os
import sys
from pathlib import Path


def find_project_root(start: Path) -> Path:
    for candidate in (start, *start.parents):
        if (candidate / "pyproject.toml").is_file() and (candidate / "fighter_rl").is_dir():
            return candidate
    raise FileNotFoundError("fighter-rl project root not found")


ROOT = find_project_root(Path.cwd().resolve())
CONFIG_PATH = ROOT / "configs/sac_lstm.json"

os.chdir(ROOT)
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
os.environ["CONFIG"] = str(CONFIG_PATH)
os.environ.setdefault("PYTHONUNBUFFERED", "1")

config = json.loads(CONFIG_PATH.read_text(encoding="utf-8"))
print(f"project: {ROOT}")
print(f"config:  {CONFIG_PATH}")
print(f"device:  {config.get('device', 'cuda')}")
print(f"output:  {config.get('output')}")
print(f"stages:  {config.get('start_stage')} -> {config.get('stop_stage')}")
```
